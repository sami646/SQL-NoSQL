```python
# TODO 3: In the same db create a new table called tomatch with two variables id_ and rdm_float.
import sqlite3
import pandas as pd

conn = sqlite3.connect(r"C:\Users\Samira\Documents\GitHub\NoSQL\data\Chap1\bc.db")
c = conn.cursor()
# le virgolette per i nomi delle colonne non sono obbligatori, li metti solo se il nome é un po strano e che potrebbec hiamare una funzione di python
def table_tomatch():
    try:
        c.execute("""CREATE TABLE tomatch("id" INT PRIMARY KEY, "rdm_float" FLOAT)""")
        conn.commit()
    except Exception as e:
        print(str(e))
```

```python
# TODO 4: In the table tomatch insert observations where the id_ starts from 50k, 
# ends at 10000000 and is only even numbers. 
#rdm_float is a float random for each observation -------------> random.random()
import tqdm

#Ho usato il code qui sotto per eliminare il contenuto della colonna id perché avevo fatto un casino
#query = "UPDATE tomatch SET id = NULL"
#c.execute(query)
#conn.commit()
#conn.close()

conn = sqlite3.connect(r"C:\Users\Samira\Documents\GitHub\NoSQL\data\Chap1\bc.db")
c = conn.cursor()

#executemany make it faster but it needs the data in a liste all'ora che execute non necessita la liste ma ci mette un botto
query = """INSERT INTO tomatch("id") VALUES (?)"""
list_of_insertion = []
for i in tqdm.tqdm(range(50000,10000000,2)):
    values = (i,)
    list_of_insertion.append(values)
    if len(list_of_insertion) == 10000:
        c.executemany(query, list_of_insertion)
        conn.commit()
        list_of_insertion = []
```
